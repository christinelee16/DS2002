```python
# aggregation pipeline that calculates average IMDb rating for each genre
avg_rating_by_genre = collection.aggregate([
    {"$unwind":"$genres"},
    {"$group":{"_id":"$genres", "avg_rating":{"$avg":"$imdb.rating"}}},
    {"$sort":{"average_rating":-1}},
    {"$limit":5}
])
for each_genre in avg_rating_by_genre:
    pprint.pprint(each_genre)
```

```python
# aggregation pipeline to find top 5 directors by avg IMDb rating of their movies
top_directors = collection.aggregate([
    {"$group":{"_id":"$directors", "avg_rating":{"$avg":"$imdb.rating"}}},
    {"$sort":{"avg_rating":-1}},
    {"$limit":5}
])
for each_director in top_directors:
    pprint.pprint(each_director)
```

```python
# aggregation pipeline to calculate total number of movies released each year
movies_per_year = collection.aggregate([
    {"$group":{"_id":"$year","total_movies":{"$sum":1}}},
    {"$sort":{"_id":1}},
])
for total_movies in movies_per_year:
    pprint.pprint(total_movies)
```

```python
# update IMDb rating of "The Godfather" to 9.5
collection.update_one({"title":"The Godfather"}, {"$set":{"imdb.rating":9.5}})
```

```python
# update all movies where genre is "horror" and set IMDb rating to 6.0 if currently null
collection.update_many({"genres":"Horror","imdb.rating":{"$exists":False}},{"$set":{"imdb.rating":6.0}})
```

```python
# delete all movies released before 1950
collection.delete_many({"year":{"$lt":1950}})
```

```python
# ensure title field is indexed for text search in MongoDB and search for movies that have "love" in title
index_info = collection.index_information()
if "title_text" not in collection.index_information():
  collection.create_index([("title","text")])

love_movies=collection.find({"$text":{"$search":"love"}})
for each_movie in love_movies:
    pprint.pprint(each_movie)
```

```python
index_info = collection.index_information()
if not any(index for index in index_info if all(field in index_info[index]['key'] for field in [('title', 'text'), ('plot', 'text')])):
  collection.create_index([("title","text"),("plot","text")])

war_movies = collection.find({"$text":{"$search":"war"}}).sort("imdb.rating",-1).limit(5)
for each_movie in war_movies:
    pprint.pprint(each_movie)
```

```python
# all movies where genre is "action" and IMDb rating is > 8
action_high_rated_movies = collection.find({"genres":"Action","imdb.rating":{"$gt":8}}).sort("year",-1)
for each_movie in action_high_rated_movies:
    pprint.pprint(each_movie)
```
